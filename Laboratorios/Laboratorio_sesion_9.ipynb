## Ejercicio  de  práctica  1.

**Aseguramiento  de  la  Calidad  de  los  Datos**

**1.** Se  tiene  un  conjunto  de  datos  sobre  empleados  que  incluye  su  nombre,  edad,  salario  y fecha de ingreso. Se requiere implementar algunos pasos de aseguramiento de la calidad de los datos, como la verificación de valores faltantes y la validación de formatos de datos.

**Datos**
**Nombre:** Juan,Ana,  Pedro,María,Luis,Ana

**Edad:** 28,  34,  None,  45,  38, 34.

**Salario:** 3000,  4000,  cinco  mil,  4500,  4000,  5000

**Fecha Ingreso:** 2021-01-15, 2020/03/12,  2022-07-01,2021/12/01,2021-05-20, 2020-03-12

**Se requiere:**

**1.** Identificar  y  contar  los  valores  faltantes  en  cada  columna.

**2.** Verificar el formato de la columna 'Salario' para asegurar que todos los valores sean numéricos.

**3.** Validar  y  estandarizar  el  formato  de  la  columna  'Fecha  Ingreso'  a  YYYY-MM-DD.

**Solución:**

```python
import pandas as pd

#Original data
data = {
    'Nombre': ['Juan', 'Ana', 'Pedro', 'Maria', 'Luis', 'Ana'],
    'Edad': [28, 24, None, 45, 38, 34],
    'Salario': ['3000', '4000', 'cinco mil', '4500', '4000', '5000'],
    'Fecha Ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12']
}

df = pd.DataFrame(data)
print("\nOriginal DataFrame:")
df

# Identificar y contar los valores faltantes
missing_values = df.isnull().sum()
print("\nValores faltantes por columna:")
print(missing_values)

# Incorrect salary
incorrect_salary = df[~df['Salario'].str.replace('[^0-9]', '',
regex=True).fillna('').str.isdigit()]
print("\nSalarios con formato incorrecto:")
print(incorrect_salary)

# Date standardization YYYY-MM-DD
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'],
errors='coerce', format='%Y-%m-%d')
print("\nDatos después de estandarizar 'Fecha Ingreso':")
print(df[['Nombre', 'Fecha Ingreso']])
```


**2.** Se tiene un conjunto de datos sobre empleados que incluye su nombre, edad, salario, puesto de trabajo y fecha de ingreso.
```python
import pandas as pd
# Cargar el DataFrame
data = {
    'Nombre': ['Juan', 'Ana', 'Pedro', 'María', 'Luis', 'Carla'],
    'Edad': [28, 24, None, 45, 38, 34],
    'Salario': ['3000', '4000', '4500', 'cuatro mil', '5000', '4000'],
    'Cargo' : ['Analista', 'Gerente', 'Desarrollador', 'Gerente', 'Analista', None],
    'Fecha Ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', None, '2020-03-12']
}
df = pd.DataFrame(data)
df
# Identify and count missing values
missing_values = df.isnull().sum()
print("\nValores faltantes por columna:")
print(missing_values)

# Incorrect salary
incorrect_salary = df[~df['Salario'].str.replace('[^0-9]', '',
regex=True).fillna('').str.isdigit()]
print("\nSalarios con formato incorrecto:")
print(incorrect_salary)

# Date standardization YYYY-MM-DD
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'],
errors='coerce', format='%Y-%m-%d')
print("\nDatos después de estandarizar 'Fecha Ingreso':")
print(df[['Nombre', 'Fecha Ingreso']])

# Missing wages
missing_wages = df[df['Cargo'].isnull()]
print("\nCargos sin valores:")
print(missing_wages)
```

```python
import pandas as pd
# Cargar el DataFrame
Data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D'],
    'Precio': [150, -25, 100, None],
}

df = pd.DataFrame(Data)
print("\nDataFrame original:")
df
# Filter products with invalid prices
invalid_prices = df[(df['Precio'] < 0) | (df['Precio'].isnull())]
print("\nProductos con precios inválidos:")
print(invalid_prices)
```

**3.** **Se requiere tener seguridad en los datos:** Encriptación y Desencriptación de Datos Sensibles,  se  tiene  un  número  de  tarjeta  de  crédito  que  deseas  encriptar  para  protegerlo y luego desencriptarlo para el uso.

**Se requiere:**

**1.** Crear una cadena con  datos sensibles (un número de tarjeta  1234-5678-9101- 1121).

**2.** Encriptar  esos  datos  utilizando  una  clave  generada.

**3.** Desencriptar  los  datos  y  mostrar  ambos  (encriptado  y  desencriptado).

**Resultado  esperado:** Ver  el  número  de  tarjeta  encriptado  y  luego  desencriptado, demostrando cómo proteger la información sensible.

**Solución:**

pip install cryptography

```python
import pandas as pd
from cryptography.fernet import Fernet

# Generar una clave
key = Fernet.generate_key()
cipher_suite = Fernet(key)

# Cargar el DataFrame
data = {
    'Nombre': ["Ana López",	"Luis Gómez",	"María Ruiz",	"Carlos Pérez",	"Lucía Torres",	"Jorge Martínez",	"Sandra Ramírez",	"David Herrera",	"Paula Díaz",	"Fernando Silva",	"Isabel Méndez",	"Tomás Ortega",	"Claudia Castro",	"Raúl Vargas",	"Diana Morales"],
    'Numero tarjeta': ["1234-5678-9012-3456",	"2345-6789-0123-4567",	"3456-7890-1234-5678",	"4567-8901-2345-6789",	"5678-9012-3456-7890",	"6789-0123-4567-8901",	"7890-1234-5678-9012",	"8901-2345-6789-0123",	"9012-3456-7890-1234",	"0123-4567-8901-2345",	"1234 5678 9012 3456",	"2345 6789 0123 4567",	"3456 7890 1234 5678",	"4567 8901 2345 6789"],
    'Fecha de vencimiento': ["12/25",	"11/24",	"10/23",	"09/22",	"08/21",	"07/20",	"06/19",	"05/18",	"04/17",	"03/16",	"02/15",	"01/14",	"12/13",	"11/12",	"10/11"],
    'CVV': ["123",	"456",	"789",	"012",	"345",	"678",	"901",	"234",	"567",	"890",	"123",	"456",	"789",	"012",	"345"],
}

df = pd.DataFrame(data)

def encrypt_field(value):
    return cipher_suite.encrypt(value.encode()).decode()

# Aplicar la función de cifrado a las columnas deseadas
df['Numero tarjeta'] = df['Numero tarjeta'].apply(encrypt_field)
df['Fecha de vencimiento'] = df['Fecha de vencimiento'].apply(encrypt_field)
df['CVV'] = df['CVV'].apply(encrypt_field)

# Mostrar el DataFrame original
print("DataFrame original:")
print(df)
```
